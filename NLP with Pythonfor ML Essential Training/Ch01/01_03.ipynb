# NLP Basics: Reading in text data & why do we need to clean the text?

### Read in semi-structured text data

```python
# Read in the raw text
import codecs
fileObj = codecs.open("SMSSpamCollection.tsv", "r", "utf_8_sig")
rawData = fileObj.read()
fileObj.close()

# Print the raw data
rawData[0:500]
```

```python
parsedData = rawData.replace('\t', '\n').split('\n')
```

```python
parsedData[0:5]
```

```python
labelList = parsedData[0::2]
textList = parsedData[1::2]
```

```python
labelList[0:5]
```

```python
textList[0:5]
```

```python
import pandas as pd

fullCorpus = pd.DataFrame({
    'label': labelList,
    'body_list': textList
})
fullCorpus.head()
```
