# NLP Basics: Exploring the dataset

### Read in text data

```python
import pandas as pd

fullCorpus = pd.read_csv('SMSSpamCollection.tsv', sep='\t', header=None)
fullCorpus.columns = ['label', 'body_text']

fullCorpus.head()
```

### Explore the dataset

```python
# What is the shape of the dataset?
print(f"Input data has {len(fullCorpus)} rows and {len(fullCorpus.columns)} columns")
```

```python
# How many spam/ham are there?
length = len(fullCorpus)
spam_len = len(fullCorpus[fullCorpus['label'] == 'spam'])
ham_len = len(fullCorpus[fullCorpus['label'] == 'ham'])

print(f"Out of {length} rows, {spam_len} are spam, {ham_len} are ham")
```

```python
# How much missing data is there?
label_nulls = fullCorpus['label'].isnull().sum()
text_nulls = fullCorpus['body_text'].isnull().sum()

print(f'Number of nulls in label: {label_nulls}')
print(f'Number of nulls in text: {text_nulls}')
```
